```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
students  = pd.read_csv('students_data.csv',  index_col=0)
students.head()
```

```python
students.columns
students.info()
```

```python
print(students.duplicated().value_counts())
students[students.duplicated(keep = False)].sort_values(by = 'admission number')  # find and count duplicate values , the second part displays only the duplicates
```

```python
print(students.duplicated(subset= ['admission number']).value_counts()) #find and count duplicate values on admission values , the second part displays only the duplicates
students[students.duplicated(subset= ['admission number'], keep = False)].sort_values(by = 'admission number') 
```

```python
 students.isna()# Returns a matrix of boolean values, where all cells containing NaN are converted to True, and all cells containing valid data are converted to False
```

```python
students.isna().sum() # since True is equivalent to 1 and False is equivalent to 0 in Python, taking the .sum() of the DataFrame (or Series) will return the total number of NaN values in the dataset.
```

```python
students.isnull().sum().sum() # total null values in a data set
```

```python
students.isnull().sum().sort_values(ascending =  False) # alternative to isna
```

## FILLING IN NULL VALUES

```python
students2 = students.fillna(value = 0) ## using the fillna function
students2.isna().sum()
```

## FILLING IN NULL VALUE WITH A PREVIOUS VALUE

```python
students3 =  students.fillna(method = 'pad')
students3.isnull().sum()

```

## filling null value with the next value row wise

```python
students4 =  students.fillna(method = 'bfill')
students4.isnull().sum()
```

## filling using the values from the next column so far we've been using row

```python
students5 =  students.fillna(method = 'bfill', axis=1)
students5.isnull().sum()
```

# Filling different values for null from different columns

```python
students6=  students.fillna({'house' : 'Nairobi',
                             'balance' : 0
                            })
students6.isnull().sum()
```

## FILLING MISSING VALUE WITH MEAN OF A COLUMN

```python
 students7 = students.fillna(value = students['english'].mean()) # you can replace this with std or min or max etc
students7.isna().sum()
```

# FILLING MISSING VALUE FROM SPECIFIC COLUMNS WITH FILLNA FUNCTION

```python
students16 = students['house'].fillna(value = 'nakuru')
students16.isna().sum()
```

## DROPPING A NULL VALUE

```python
students8  =  students.dropna()
students8.isna().sum()
```

```python
students9  =  students.dropna(how = 'all') # drops row only if all values are null
students9.isna().sum()
```

```python
students10  =  students.dropna(how = 'any') # drops row  if any of the  values are null
students10.isna().sum()
```

```python
students.describe()
```

## replace

```python
students11 = students.replace(to_replace=np.nan, value= 98) ## NOT RESTRICTED TO NAN CAN REPLACE ANY VALUE ON THE DATA SET LIKE THIS students.replace(to_replace=100.0, value= 99)
students11.isna().sum()
```

```python
students11.describe()
```

## Using interpolate()

```python
students['balance'] =  students['balance'].interpolate( method= 'linear') ## does mean between the two actual values and supplies it to the null value in the middle

students['balance'].isna().sum()
```
