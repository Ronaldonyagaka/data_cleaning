```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
students  = pd.read_csv('students_data.csv',  index_col=0)
students.tail()
```

```python
students.columns
students.info()
```

```python
print(students.duplicated().value_counts())
students[students.duplicated(keep = False)].sort_values(by = 'admission number')  # find and count duplicate values , the second part displays only the duplicates
```

```python
print(students.duplicated(subset= ['admission number']).value_counts()) #find and count duplicate values on admission values , the second part displays only the duplicates
students[students.duplicated(subset= ['admission number'], keep = False)].sort_values(by = 'admission number') 
```

```python
students.isna() # Returns a matrix of boolean values, where all cells containing NaN are converted to True, and all cells containing valid data are converted to False
```

```python
students.isna().sum() # since True is equivalent to 1 and False is equivalent to 0 in Python, taking the .sum() of the DataFrame (or Series) will return the total number of NaN values in the dataset.
```
